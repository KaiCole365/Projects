# **My First ML Project**

# **Load Data**

```python
import pandas as pd

# Import the delaney solubility csv file
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df

# Y variable = logS
# X variables = the rest of the data in the columns
```

# **Data Preparation**

### Data Separation as X and Y

```python
# Filter for y variable
y = df['logS']
y
```

```python
# Filter for X variables
x = df.drop('logS', axis=1)
x
```

### Data splitting

```python
# Import sklearn and train test split for train and testing models
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train

# 80% of data used to train model
```

```python
x_test

# 20% of data used to test model
```

# **Model Building** **bold text**

## Linear Regression

**Training the model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

###**Applying the model to make a prediction**

```python
# Creating model to make a prediction on x training model
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred

# All of the predictions representing 80% of the data
```

```python
y_lr_test_pred

# All predictions representing 20% of the data
```

###**Evaluate model performance**

Comparing the predictive model (y_lr_train_pred) to the actual data (y_train)

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculating the Mean Squared Error for training set of data
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

# Calculating the Mean Squared Error for the testing set of data
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
```

```python
lr_results
```

```python
# Renaming the columns 0-4
lr_results.columns = ['Method', 'Trainin MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

# **Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

## **Applying the model to make a prediction**

```python
# Creating model to make a prediction on x training model
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

## **Evaluate model performance**

```python
# Calculating the Mean Squared Error for training set of data
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

# Calculating the Mean Squared Error for the testing set of data
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()

# Renaming the columns 0-4
rf_results.columns = ['Method', 'Trainin MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **Model Comparison**

```python
# Combining and comparing the tables of the two methods (LR vs RF)
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
# Fixing the index numbers in the 'Method' column
df_models.reset_index(drop=True)
```

# **Data Visualization of Prediction Results**

```python
# Importing matplotlib.pyplot to graph data via scatter plot
import matplotlib.pyplot as plt

# Impoting numpy to create a fitted line on the graph
import numpy as np


# Set the width and height of graph
plt.figure(figsize=(5,5))

# Create scatter plot graph of trained data
plt.scatter(x=y_train, y=y_lr_train_pred, c="green", alpha=0.3)

# Create the fitted line on the graph
z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), 'red')

# Label the X and Y axis
plt.ylabel('Predict LogS')
plt.xlabel('Experimental LogS')
plt.plot()
```
