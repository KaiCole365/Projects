# Cleaning the Data

### Goal: To find the strongest Marvel Character

```python
# Filtering for relevant data
df1 = df.filter(['Name', 'Strength', 'Power', 'Speed', 'Combat'])
df1
```

```python
# Calculating total NULL values
df1.isnull().sum()
```

```python
# Dropping all NULL values in the dataset
df2 = df1.dropna()
df2.isnull().sum()
```

```python
df2.head(10)
```

```python
# Calculating a combined strength score
df2 = df2.copy()
df2['Strength_Score'] = df2['Strength'] + df2['Power'] + df2['Speed'] + df2['Combat']
```

```python
# Plotting the top 10 strongest Marvel Characters
top_10 = df2.nlargest(10, 'Strength_Score')

plt.figure(figsize=(12, 6))

# Creating a stacked barplot to graph data
bottom = np.zeros(10)
for attribute in ['Strength', 'Power', 'Speed', 'Combat']:
    plt.bar(top_10['Name'], top_10[attribute], bottom=bottom, label=attribute)
    bottom += top_10[attribute]

plt.title('Strongest Marvel Character')
plt.xlabel('Name')
plt.ylabel('Combined Strength Score')
plt.legend()
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
                            
```

## Just for fun, let's see the breakdown of Marvel Character Alignment

```python
df3 = df.filter(['Name', 'Alignment_x'])
df3.head(5)
```

```python
# Renaming 'Alignment_x'
df3.rename(columns={'Alignment_x' : 'Alignment'}, inplace=True)
df3.head(5)
```

```python
# Plotting Marvel Character Alignment via Pie Chart
alignment_counts = df3['Alignment'].value_counts()

plt.figure(figsize=(10, 6))
wedges, texts, autotexts = plt.pie(alignment_counts.values,
                                   labels=alignment_counts.index,
                                   autopct='%1.1f%%',
                                   pctdistance=0.85)

# Moving labels outside the pie for clarity
plt.setp(texts, size=8, weight="bold")
plt.setp(autotexts, size=8, weight="bold")

plt.title('Marvel Character Alignment Distribution')
plt.axis('equal')

# Adding legend for added detail
plt.legend(wedges, alignment_counts.index,
           title="Alignments",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```
